# Plot Results of training
allows interpretation of impact of various parameters setting 

### needs correct <file>.csv file

```python
import numpy as np
import pandas as pd
import csv
```

```python
# matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# which file is the source

file = 'trial_may31.csv'
```

```python
with open(file, 'rb') as csvfile:
    reader = csv.DictReader(csvfile)
    fscore = []
    val_accuracy = []
    train_loss = []
    val_loss = []
    x = []
    for row in reader:
        
        fscore.append([float(row[' Val_fscore'])])
        val_accuracy.append([float(row[' Val_acc'])])
        train_loss.append([float(row[' Train_loss'])])
        val_loss.append([float(row[' Val_loss'])])
        x.append(row['Epoch'])
```

```python
# plotting

fig = plt.figure()
fig.set_size_inches(18.5, 10.5)

ax = fig.add_subplot(1,1,1)
ax.set_title('Trining Results')
ax.set_xlabel('# Epochs')
ax.set_ylabel('value')


ax.plot(fscore, label='fscore');
ax.plot(val_accuracy, label='val_acc');
ax.plot(train_loss, label='train_loss');
ax.plot(val_loss, label='val_loss');
legend = ax.legend(loc='upper left', shadow=False)
```
